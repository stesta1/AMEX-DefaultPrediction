```python
from sklearn.utils.random import sample_without_replacement
import numpy as np
import pandas as pd
```

```python
path_train="path/amex-default-prediction/train_data.csv"
path_label='path/amex-default-prediction/train_labels.csv'
```

```python
df_train=pd.read_csv(path_train)
df_labels=pd.read_csv(path_label)
```

```python
df_labels.head()
```

```python
df_train.head()
```

```python
numero_grupo=14
random_state=(31416 * numero_grupo) % 1000
```

```python
total=df_train.shape[0]#5531451
muestra=np.round(total*5/100)
samples=sample_without_replacement(total,muestra, random_state=random_state)
```

```python
df_samples=df_train.iloc[samples]
```

```python
df_samples
```

```python
df=pd.merge(df_samples,df_labels,on='customer_ID',how='left')
```

```python
df.target.value_counts()
```

```python
#df.reset_index() 
df.to_csv("path/train_sampled_grupo14.csv",index=False,header=True)
```
